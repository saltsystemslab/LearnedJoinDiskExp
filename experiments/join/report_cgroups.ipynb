```python
import os;
import json;
import builtins
import duckdb
import pandas as pd;
import matplotlib.pyplot as plt

def get_dataset_from_testcase(run):
    tokens = run.split("_")
    return tokens[0]

def get_memlim_from_testcase(run):
    tokens = run.split("_")
    return tokens[1][7:]

def get_cgroup_lim_from_testcase(run):
    if run.find('4G') != -1:
        return 4024
    if run.find('2G') != -1:
        return 2048
    if run.find('256M') != -1:
        return 256
    if run.find('512M') != -1:
        return 512
    return 0

dirs = [
    '/home/chesetti/Repos/LearnedJoinDiskExp/sponge/join_unsorted_cgroup',
    '/home/chesetti/Repos/LearnedJoinDiskExp/sponge/join_unsorted_cgroup_4G',
    '/home/chesetti/Repos/LearnedJoinDiskExp/sponge/join_unsorted_cgroup_2G_run2',
]
runs = []
for dir in dirs:
 for test_case in os.listdir(dir):
    if test_case == 'build':
        continue
    for run_name in os.listdir(os.path.join(dir, test_case, 'outputs', 'results')):
        if (run_name=="run"):
           continue
        if ('sanity' in run_name):
            continue
        json_file = builtins.open(os.path.join(dir, test_case, 'outputs', 'results',  run_name))
        run = (json.load(json_file))
        run['dataset'] = get_dataset_from_testcase(test_case)
        run['memlimit'] = get_cgroup_lim_from_testcase(dir)
        print(dir)
        runs.append(run)
df = pd.json_normalize(runs)
print(df.columns)
display(df)
```

```python

epsilon = {
    "pgm256": 256,
    "pgm1024": 1024,
    "pgm2048": 2048,
    "flatpgm256": 256,
    "flatpgm1024": 1024,
    "flatpgm2048": 2048,
    "flatpgm4096": 4096,
    "flatpgm8192": 8192,
    "hashjoin": 0,
    "btree": 0,
    "sj": 0,
    "sj2": 0,
    "standard_merge": 0
};
def indexType(x):
    if "pgm" in x:
        return "PGM"
    elif "btree" in x:
        return "BTREE"
    return "NA"

def getRatio(tableName):
    print(tableName)
    if 'table1000' in tableName:
        return 1000
    if 'table100' in tableName:
        return 100
    if 'table10' in tableName:
        return 10
    if 'table1' in tableName:
        return 1
    return -1

df["threads"] = df["spec.num_threads"]
df["duration_sec"] = df["result.duration_ns"] / (1000000000)
df['result_num_keys'] = df['result.num_output_keys']
df["ratio"] = df['spec.outer_table'].map(lambda x: getRatio(x))
df["algo"] = df["spec.algo"]
df["index_type"] = df["spec.algo_name"].map(lambda x: indexType(x))
df['index_creation_sec'] = df['result.inner_index_duration_ns']/1e9; 
df['actual_join_duration_sec'] = df['result.join_duration_ns']/1e9; 
df['dataset'].unique()
df['result.num_output_keys'].unique()
display(duckdb.sql("SELECT * FROM df WHERE dataset='fb'").df())
```

```python
def print_for_dataset(algos, memlim):
    print(memlim)
    rows = duckdb.sql(
        f" SELECT memlimit, ratio, algo, actual_join_duration_sec as v FROM df WHERE memlimit={memlim} GROUP BY memlimit, ratio, algo, actual_join_duration_sec"
    ).df()
    rows = rows.pivot(index='ratio', values='v', columns='algo')
    display(rows)
    rows[algos].to_csv('/home/chesetti/Repos/LearnedJoinDiskExp/experiments/results/join_unsorted_cgroup/'+str(memlim)+'.txt')

def print_for_dataset_index_build_duration(algos, memlim):
    print(memlim)
    rows = duckdb.sql(
        f" SELECT memlimit, ratio, algo, index_creation_sec as v FROM df WHERE memlimit={memlim} AND ratio=1"
    ).df()
    print(rows)
    rows = rows.pivot(index='memlimit', values='v', columns='algo')
    print(rows.to_latex())




algos = ['unsorted_lsj_sorted_output', 'unsorted_inlj_sorted_output']
print_for_dataset(algos, 4024)
print_for_dataset(algos, 2048)
print_for_dataset(algos, 0)

print_for_dataset_index_build_duration(algos, 4024)
print_for_dataset_index_build_duration(algos, 2048)
print_for_dataset_index_build_duration(algos, 0)
```

```python
algos = ['lsj_sampledflatpgm256', 'lsj_flatpgm256', 'lsj_radixspline256', 'lsj_rmi' , 'lsj_btree256', 'sort_join']
print_for_dataset("fb", algos)
print_for_dataset("wiki", algos)
print_for_dataset("osm", algos)
print_for_dataset("books", algos)
```
